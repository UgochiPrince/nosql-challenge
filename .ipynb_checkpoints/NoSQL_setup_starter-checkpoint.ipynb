2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields

# Define the query to find the BusinessTypeID for "Restaurant/Cafe/Canteen"
query = {
    "BusinessType": "Restaurant/Cafe/Canteen"
}

# Project only the BusinessTypeID and BusinessType fields
projection = {
    "_id": 0,  # Exclude the default "_id" field
    "BusinessTypeID": 1,
    "BusinessType": 1
}

# Find the document(s) matching the query and project the specified fields
result = establishments.find(query, projection)

# Loop through the results and print each document
for doc in result:
    pprint(doc)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID

# Define the query to find the BusinessTypeID for "Restaurant/Cafe/Canteen"
business_type_query = {
    "BusinessType": "Restaurant/Cafe/Canteen"
}

# Project only the BusinessTypeID field
business_type_projection = {
    "_id": 0,  # Exclude the default "_id" field
    "BusinessTypeID": 1
}

# Find the document(s) matching the query and project the BusinessTypeID
business_type_result = establishments.find_one(business_type_query, business_type_projection)

if business_type_result:
    # Extract the BusinessTypeID from the result
    correct_business_type_id = business_type_result["BusinessTypeID"]

    # Define the query to update the new restaurant
    update_query = {
        "BusinessName": "Penang Flavours"
    }

    # Set the correct BusinessTypeID in the update
    update_values = {
        "$set": {
            "BusinessTypeID": correct_business_type_id
        }
    }

    # Update the document
    update_result = establishments.update_one(update_query, update_values)
    
    # Confirm that the new restaurant was updated
    if update_result.modified_count > 0:
        print("The new restaurant's BusinessTypeID has been updated.")
    else:
        print("No document was updated. Please check the query.")
else:
    print("BusinessTypeID for 'Restaurant/Cafe/Canteen' not found.")
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"

# Define the query to find documents with LocalAuthorityName as "Dover"
query = {
    "LocalAuthorityName": "Dover"
}

# Count the documents that match the query
document_count = establishments.count_documents(query)
print(f"Number of documents with LocalAuthorityName 'Dover': {document_count}")
```

```python
# Delete all documents where LocalAuthorityName is "Dover"

# Define the query to find documents with LocalAuthorityName as "Dover"
query = {
    "LocalAuthorityName": "Dover"
}

# Delete all documents that match the query
delete_result = establishments.delete_many(query)

# Print the number of documents deleted
print(f"Number of documents deleted: {delete_result.deleted_count}")
```

```python
# Check if any remaining documents include Dover

# Define the query to find documents with LocalAuthorityName containing "Dover"
query = {
    "LocalAuthorityName": {
        "$regex": "Dover",
        "$options": "i"  # Case-insensitive search
    }
}

# Count the documents that match the query
document_count = establishments.count_documents(query)

if document_count > 0:
    print("There are remaining documents with 'Dover' in LocalAuthorityName.")
else:
    print("No remaining documents include 'Dover' in LocalAuthorityName.")
```

```python
# Check that other documents remain with 'find_one'

# Define a query to find one document that doesn't have 'Dover' in LocalAuthorityName
query = {
    "LocalAuthorityName": {
        "$ne": "Dover"  # "$ne" means "not equal to"
    }
}

# Find and display one document that matches the query
remaining_document = establishments.find_one(query)

if remaining_document:
    print("There are remaining documents in the collection.")
    # Print the first remaining document
    pprint(remaining_document)
else:
    print("No remaining documents found.")

```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Define the update queries to convert geocode.longitude and geocode.latitude
longitude_query = {"geocode.longitude": {"$type": "string"}}
latitude_query = {"geocode.latitude": {"$type": "string"}}

# Use $toDouble to convert geocode.longitude and geocode.latitude
longitude_update = [{"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"}}}]
latitude_update = [{"$set": {"geocode.latitude": {"$toDouble": "$geocode.latitude"}}}]

# Apply the updates using update_many
update_result = establishments.bulk_write([
    pymongo.UpdateMany(longitude_query, longitude_update),
    pymongo.UpdateMany(latitude_query, latitude_update)
])

# Print the number of documents updated
print(f"Number of documents updated: {update_result.modified_count}")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

# Define the update query to convert RatingValue
rating_value_query = {"RatingValue": {"$type": "string"}}

# Use $toInt to convert RatingValue to an integer
rating_value_update = [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}]

# Use update_many to update multiple documents
update_result = establishments.bulk_write([
    pymongo.UpdateMany(rating_value_query, rating_value_update)
])

# Print the number of documents updated
print(f"Number of documents updated: {update_result.modified_count}")
```

```python
# Check that the coordinates and rating value are now numbers

# Check the data type of geocode.longitude, geocode.latitude, and RatingValue fields
example_document = establishments.find_one()

if example_document:
    longitude_type = type(example_document.get("geocode", {}).get("longitude"))
    latitude_type = type(example_document.get("geocode", {}).get("latitude"))
    rating_value_type = type(example_document.get("RatingValue"))

    print(f"Data type of geocode.longitude: {longitude_type}")
    print(f"Data type of geocode.latitude: {latitude_type}")
    print(f"Data type of RatingValue: {rating_value_type}")
    
else:
    print("No document found in the collection.")
```
